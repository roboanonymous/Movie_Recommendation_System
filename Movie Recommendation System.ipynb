```python
moviemat = df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
```

```python
moviemat
```

```python
rating.sort_values('num of rating', ascending = False).head()
```

```python
starwars_movierating = moviemat['Star Wars (1977)']
```

```python
similarto_starwars = moviemat.corrwith(starwars_movierating)
similarto_starwars
```

```python
corr_starwars = pd.DataFrame(similarto_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

```python
corr_starwars.sort_values(by='Correlation',ascending=False).head(10)
```

```python
corr_starwars[corr_starwars['num of rating']>100].sort_values('Correlation',ascending=False)
```

```python
def predict_movies(moviename):
    movie_user_ratings = moviemat[moviename]
    similar_to_movie = moviemat.corrwith(movie_user_ratings)
    
    corr_movie = pd.DataFrame(similar_to_movie,columns=['Correlation'])
    corr_movie.dropna(inplace=True)
    
    corr_movie = corr_movie.join(rating['num of rating'])
    predictions = corr_movie[corr_movie['num of rating'] > 100].sort_values('Correlation',ascending=False)
    return predictions
```

```python
predictions = predict_movies("Titanic (1997)")
```
